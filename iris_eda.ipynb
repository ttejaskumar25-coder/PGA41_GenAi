# Iris Dataset Exploratory Data Analysis

This notebook performs an exploratory data analysis (EDA) on the Iris dataset located in `data/Iris.csv`. The goal is to understand the dataset structure, feature distributions, species relationships, and draw concise conclusions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

iris_path = 'data/Iris.csv'
iris = pd.read_csv(iris_path)
iris.head()
```

## Dataset Overview

Review the dataset structure, columns, and sample records. Confirm whether there are missing values and identify the target variable.

```python
iris.info()

iris.isna().sum()
```

## Summary Statistics

Generate descriptive statistics for all numeric features to understand central tendency, spread, and range.

```python
iris.describe()
```

## Species Distribution

Inspect how observations are distributed across the three Iris species. Check for class balance.

```python
iris['Species'].value_counts()

plt.figure(figsize=(8, 5))
sns.countplot(data=iris, x='Species')
plt.title('Iris Species Count')
plt.xlabel('Species')
plt.ylabel('Count')
plt.tight_layout()
```

## Feature Distributions by Species

Visualize the distribution of each numeric feature and compare how measurements differ by species.

```python
feature_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
iris_melt = iris.melt(id_vars='Species', value_vars=feature_cols, var_name='Feature', value_name='Value')

plt.figure(figsize=(12, 10))
sns.boxplot(data=iris_melt, x='Feature', y='Value', hue='Species')
plt.title('Feature Distributions by Species')
plt.legend(loc='upper right')
plt.tight_layout()
```

## Pairwise Relationships and Correlation

Analyze pairwise relationships between numeric features and compute the correlation matrix.

```python
plt.figure(figsize=(10, 8))
sns.pairplot(iris, hue='Species', vars=feature_cols, corner=True)
plt.suptitle('Pairwise Feature Relationships by Species', y=1.02)

corr = iris[feature_cols].corr()
corr
```

## Conclusions

- The dataset contains 150 observations and 5 columns (including the species label).
- All species are balanced with 50 samples each, making this dataset well-suited for classification.
- Petal length and petal width show the strongest separation between species, especially for `Iris-setosa` vs others.
- Sepal measurements display more overlap across species, so they are less discriminative on their own.
- Strong positive correlation exists between petal length and petal width, while sepal width is less correlated with sepal length.

These findings indicate that petal features are the most informative for species classification in the Iris dataset.
